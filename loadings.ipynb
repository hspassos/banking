```python
import pandas as pd

cosif = pd.read_csv('Códigos COSIF.csv', sep=';', encoding='utf-8', header=3)

cosif['conta'] = cosif['CÓDIGOS'].str.replace(r'[.-]', '', regex=True)

# Converte para número. O que for lixo vira NaN
cosif['conta'] = pd.to_numeric(cosif['conta'], errors='coerce')

# Opcional: Remover as linhas que viraram NaN
cosif = cosif.dropna(subset=['conta'])

# Agora convertemos para int (após limpar os NaNs)
cosif['conta'] = cosif['conta'].astype(int)

cosif.to_csv('cosif.csv', index=False)
```
